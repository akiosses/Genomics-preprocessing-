```python
import pandas as pd
import math 

df = pd.read_excel('heart failure_Final_table.xlsx')

df.index=[x for x in range (1,len(df.values)+1)]
df.index.name = 'id'

df
```

```python
#from "sample" choose the Europeans 

to_print = df['sample'].unique()
#print(to_print)
df2=df[~df["sample"].str.contains('(African|Mixed|American)', 
regex=True)]
df2
```

```python
#clean unnecessary symbols 

unwanted="+-.]['"

for char in unwanted:
 df['Direction']=df['Direction'].str.replace(char,"")
 df['title_pubmedId']=df['title_pubmedId'].str.replace(char,"")
 df['trait_Pubmed']=df['trait_Pubmed'].str.replace(char,"")
 df['trait_SNPedia'] = df['trait_SNPedia'].str.replace(char, "")
df
```

```python
#for 2 or more Snp corresponding to the same ChrId, retain the snp with the most significant p-value

to_print2 = df.groupby(['SNP','ChrId', ]).agg({'pValue':[min]})
to_print2
```

```python
#SNPs having p-value from 10^(-7) to 10^(-4).

df_snp=df[(df.pValue >= math.pow(10,-7))&(df.pValue <= math.pow(10,-4))]
df_snp
```

```python
#create a new categorical feauture named "Significance" such that:

#Significance= HS : if 10^(-8) < p <10^(-7) 
#Significance= S : if 10^(-7) < p <10^(-5) 
#Significance= NS : if 10^(-5) < p <10^(-4) 

df['Significance']=df['pValue'].map(lambda pValue:'HS' 
if(pValue<math.pow(10,-7))&(pValue>= math.pow(10,-8)) else ('S' if 
(pValue<= math.pow(10,-5))&(pValue>= math.pow(10,-7)) else ('NS' if 
(pValue>math.pow(10,-5))&(pValue<math.pow(10,-4)) else "")))
df
```

```python
#merge the column "Significance" with the table df_snp

dfmerge2=pd.merge(df_snp,df['Significance'],on='id')
dfmerge2
#dfmerge2.to_excel("casestudymerge6.xlsx")
```
